# Add prediction columns base on category
New columns will be added for prediction columns as below:
1. The total project cost is split into three categories of (low, medium low, medium high and high) with indices of 0,1,2,3
2. Ontario commitment is split into three categories of (low, medium low, medium high and high) with  indices of 0,1,2,3
3. Additional columns Additionalfor Ontario share will be added (later will be used in regression)


1.  Create a new column as 'label total project costs' based on the values of first interquantile, median and third interquantile. 

```python
df['total_project_costs']
```

```python
df['total_project_costs'].median(),df['total_project_costs'].mean()
```

```python
x = np.quantile(df['total_project_costs'], [0,0.25,0.5,0.75,1])
x
```

```python
np.quantile(df['total_project_costs'], [0.25])[0]
```

```python
def categorise1(df):  
    if df['total_project_costs'] < 2.5688175e+05:
        return '0'
    elif df['total_project_costs'] > 2.5688175e+05 and df['total_project_costs'] <= 4.1714000e+05:
        return '1'
    elif df['total_project_costs'] > 4.1714000e+05  and df['total_project_costs'] <= 8.0642075e+05:
        return '2'
    if df['total_project_costs'] > 8.0642075e+05:
        return '3'
```

```python
df['label_total_project_costs'] = df.apply(lambda df: categorise1(df), axis=1)
```

```python
df['label_total_project_costs'].value_counts() 
```

```python
df['label_total_project_costs']=df['label_total_project_costs'].astype(int)
```

```python
df.info()
```

2. Create a new column as 'label ontario commitment' based on the values of first interquantile, median and third interquantile.

```python
x = np.quantile(df['ontario_commitment'], [0,0.25,0.5,0.75,1])
x
```

```python
def categorise2(df):  
    if df['ontario_commitment'] <= 1.0000000e+05:
        return '0'
    elif df['ontario_commitment'] > 1.0000000e+05 and df['ontario_commitment'] <= 1.5000000e+05:
        return '1'
    elif df['ontario_commitment'] > 1.5000000e+05  and df['ontario_commitment'] < 2.6999300e+05:
        return '2'
    if df['ontario_commitment'] >= 2.6999300e+05:
        return '3'
```

```python
df['label_ontario_commitment'] = df.apply(lambda df: categorise2(df), axis=1)
```

```python
df['label_ontario_commitment'].value_counts()
```

```python
df['label_ontario_commitment']=df['label_ontario_commitment'].astype(int)
```

# Create two categories

```python
x = np.quantile(df['ontario_commitment'], [0,0.25,0.5,0.75,1])
x
```

```python
def two_categories(df):  
    if df['ontario_commitment'] <= 1.4000000e+05:
        return '0'
    elif df['ontario_commitment'] > 1.4000000e+05:
        return '1'
```

```python
df['two_labela_ontario_commitment'] = df.apply(lambda df: two_categories(df), axis=1)
```

```python
df['two_labela_ontario_commitment']=df['two_labela_ontario_commitment'].astype(int)
```
